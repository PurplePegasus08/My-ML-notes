EDA -> Exploratory Data Analysis
Exploratory Data Analysis (EDA) is a simple process of understanding and exploring your data using visualizations and summary statistics. It helps find patterns, relationships between variables, correlation analysis, detect outliers, and form hypotheses before diving into more complex analyses.


EDA involves:
- Univariate: Uni means one and variate means variable, so in univariate analysis, there is only one dependable variable.
- Bivariate visualization: Bi means two and variate means variable, so here there are two variables. The analysis is related to cause and the relationship between the two variables. 
- Multivariate visualization: Multivariate analysis is required when more than two variables have to be analyzed simultaneously.

Types of Data
- Numerical Data - Data in the form of numbers. Continous values, Discrete value etc. Discrete data includes discrete variables that are finite, numeric, countable, and non-negative integers.
- Categorical Data -a type of data that can be stored into groups or categories 
- Time-series Data - Data that is recorded over consistent intervals of time.
- Ordinal Data -Ordinal data classifies data while introducing an order, or ranking.

```python
#difference between countplot and barplot
#countplot use for occurence of each variable
#barplot show the average of each variable

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
data = {'Product': ['A', 'B', 'A', 'C', 'B', 'A'],
        'Quantity': [2, 3, 1, 4, 2, 3]}
df = pd.DataFrame(data)

# Countplot: Shows the number of occurrences of each product
sns.countplot(x='Product', data=df)
plt.title('Countplot: Product Frequency')
plt.show()

# Barplot: Shows the average quantity of each product
sns.barplot(x='Product', y='Quantity', data=df)
plt.title('Barplot: Average Quantity per Product')
plt.show()
```

```python
import plotly.graph_objects as go

# Create a 3D scatter plot using Plotly
fig = go.Figure(data=[go.Scatter3d(
    x=bat['avg'],
    y=bat['strike_rate'],
    z=bat['runs'],
    mode='markers',
    marker=dict(size=5, color=bat['runs'], colorscale='Viridis', opacity=0.8)
)])

# Add axis titles
fig.update_layout(scene=dict(
                    xaxis_title='Batting Average',
                    yaxis_title='Strike Rate',
                    zaxis_title='Runs'),
                    title="Interactive 3D Batting Plot")

# Show interactive plot
fig.show()
```

```python
import plotly.express as px

# Create an interactive 2D scatter plot with Plotly
fig = px.scatter(bat, x='avg', y='runs', title="Interactive 2D Scatter Plot",
                 labels={'avg': 'Batting Average', 'runs': 'Runs'},
                 template='plotly_dark')

# Show interactive plot
fig.show()
```
